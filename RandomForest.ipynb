```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
import pylab as pl
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv("data/train.csv")
data.head()
```

```python
data = data.drop(['Id'],axis = 1)
```

```python
features = data.drop(['Cover_Type'],axis = 1)
labels = data['Cover_Type']
labels_categorical = pd.get_dummies(labels)
labels_categorical.head()

X_train,X_test,Y_train,Y_test = train_test_split(features,labels_categorical,test_size=0.20,random_state = 42)
X_test
```

```python
clf = RandomForestClassifier()
clf.fit(X_train, Y_train) 
Y_pred = clf.predict(X_test)
print(accuracy_score(Y_test,Y_pred))
```

```python
clf = RandomForestClassifier(n_estimators=107, max_depth=70,
                             random_state=0,criterion='entropy',min_samples_split=2)
clf.fit(X_train, Y_train) 
Y_pred = clf.predict(X_test)
print(accuracy_score(Y_test,Y_pred))
```
